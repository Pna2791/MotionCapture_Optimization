```python
from scipy.spatial.transform import Rotation
import numpy as np

def create_rotation(array):
    euler_angles = np.radians(array)
    r = Rotation.from_euler('xyz', euler_angles)
    rotation_matrix = r.as_matrix()
    return rotation_matrix

def Rot2Euler(rotation_matrix):
    r = Rotation.from_matrix(rotation_matrix)
    return np.degrees(r.as_euler('xyz'))
```

```python
def compute(Root, T_pose, status):
    R_Gp_B0 = np.array([
        [ 0.,  0.,  1.],
        [ 1.,  0., -0.],
        [ 0.,  1.,  0.]
    ])
    R_Gn_Gp = Root
    R_Gn_S0 = T_pose
    
    R_Gp_S0 = R_Gn_Gp.transpose().dot(R_Gn_S0)
    R_B0_S0 = R_Gp_B0.transpose().dot(R_Gp_S0)
    
    R_Gn_St = status
    R_Gp_St =  R_Gn_Gp.transpose().dot(R_Gn_St)
    R_Gp_Bt = R_Gp_St.dot(R_B0_S0.transpose())
    
    return R_Gp_Bt
```

```python
Root = create_rotation(
    [0, 0, 0]
)
T_pose = create_rotation(
    [0, 0, 90]
)
status = create_rotation(
    [0, 0, 0]
)

rot = compute(Root, T_pose, status)

print(np.round(rot, 2))
print()

print(np.round(Rot2Euler(rot)))
```

```python
target = np.array(
    [[ 1.,  0.,  0.],
 [ 0.,  0. ,-1.],
 [ 0. , 1. , 0.]], dtype=np.float32
)
target
```

```python
value_list = list(range(-180, 180, 10))


for x in value_list:
    for y in value_list:
        for z in value_list:
            rot = compute(
                create_rotation(
                    [x, y, z]
                ),
                T_pose,
                status
            )
            if np.round(rot, 1).all() == target.all():
                print(x, y, z)
```

```python
root_rot = np.array([
    [0, 0, 1],
    [1, 0, 0],
    [0, 1, 0],
], dtype=np.float32
)

euler_angles = np.radians([45, 0, 0])
r = Rotation.from_euler('xyz', euler_angles)
rotation_matrix = r.as_matrix()
rot = rotation_matrix.dot(root_rot )

print(rot)
print()

print(Rot2Euler(rot))
```
